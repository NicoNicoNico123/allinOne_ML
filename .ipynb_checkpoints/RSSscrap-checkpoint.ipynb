```python
import feedparser
import pandas as pd
```

```python
# Parse the RSS feed
d = feedparser.parse('https://www.investing.com/rss/news_301.rss')

# Extract the titles
titles = [entry.title for entry in d.entries]

# Convert the list of titles to a DataFrame
df = pd.DataFrame(titles, columns=['Title'])

# Save the DataFrame to a CSV file
df.to_csv('titles.csv', index=False)
```
